```python
#Feature Engineering
import pandas as pd

df = pd.read_csv("../data/Library_Usage.csv")
df.head()
```

```python
#Konvertiere die Spalte in ein numerisches Format
import pandas as pd
df['Circulation Active Year'] = pd.to_numeric(
  df['Circulation Active Year'], errors='coerce'
)
df['Circulation Active Year']
```

```python
#zuerst konvertieren wir die Spalte in ein Datumsformat. 
df['Circulation Active Month'] = pd.to_datetime(
    df['Circulation Active Month'],
    errors='coerce',
    format="%B"
)
df['Circulation Active Month']
```

```python
#Jetzt extrahieren wir den Monat als Zahl aus der Spalte:
df['Circulation Active Month'] = df['Circulation Active Month'].dt.month
df['Circulation Active Month']
```

```python
df['Membership Duration'] = (
    df['Circulation Active Year'] - df['Year Patron Registered']
)*12 + df['Circulation Active Month']
```

```python
df['Membership Duration'] = df['Membership Duration'].fillna(0)
df['Membership Duration']
```

```python
#Fehlende Werte
import pandas as pd
df = pd.read_csv('../data/Library_Usage.csv')
```

```python
#welche Spalten enthalten alles fehlende Werte
df.isna().sum()
```

```python
#Lies den Datensatz ein und erstelle einen df der keine Beobachtungen mit fehlenden Werten mehr enthält
df_clean = df.dropna()
df_clean.isna().sum()
```

```python
#Wie viele Beobachtungen wurden entfernt?
len(df) - len(df_clean)
```
